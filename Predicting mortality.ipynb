```r
Mortality_data <- read.csv("C:/Users/ADMIN/Downloads/USA deaths data.csv")
```

```r
head(Mortality_data)
```

```r
tail(Mortality_data)
```

```r
Males <- Mortality_data[, c(1, 3)]
Frmales <- Mortality_data[, c(1, 2)]
```

```r
colnames(Males)[colnames(Males) == "Male"] <- "dx"
```

```r
colnames(Females)[colnames(Frmales) == "Female"] <- "dx"
```

```r
# Summary statistics 
#provide a quick overview of the dataset's central tendency, dispersion, and range.
summary(Males$dx)
summary(Females$dx)
```

```r
plot(Males)
plot(Females)
```

```r
str(Males)
str(Females)
#dx column is an integer data type ,we convert to numeric so as to plot a histogram 
```

```r
# Convert the column to numeric
Males$dx<- as.numeric(Males$dx)
Females$dx<- as.numeric(Females$dx)
```

```r
str(Males)
str(Females)
```

```r
#histogram
#visualize the distribution of the data and help identify patterns, skewness, and potential outliers.
hist(Males$dx)
hist(Females$dx)
```

```r
# Boxplot
#To identify any outliers and understand the distribution better.
#summarize the distribution of the data and provide insights into central tendency, variability, and potential outliers.
boxplot(Males$dx, main = "Boxplot of Number of Male Deaths", ylab = "Number of Deaths")
boxplot(Females$dx, main = "Boxplot of Number of Female Deaths", ylab = "Number of Deaths")
```

```r
# Density plot
#Visualize the density of the number of deaths
#estimate the probability density function of the data, helping to visualize the shape and spread of the distribution more smoothly than histograms.
plot(density(Males$dx), main = "Density Plot of Number of Male Deaths", xlab = "Number of Deaths")
plot(density(Females$dx), main = "Density Plot of Number of Female Deaths", xlab = "Number of Deaths")
```

```r
# QQ plot for males 
#To check the normality assumption 
#compare the distribution of the data against a theoretical normal distribution, helping to assess the data's normality.
qqnorm(Males$dx)
qqline(Males$dx)
```

```r
# QQ plot for females
#To check the normality assumption
qqnorm(Females$dx)
qqline(Females$dx)
```
